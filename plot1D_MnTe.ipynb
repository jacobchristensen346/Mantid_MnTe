```python
import matplotlib.transforms as mtransforms


fig, (ax1,ax2) = plt.subplots(1,2,figsize=(9,3.5))
zcut = np.loadtxt('normData_MnTe_370K-330K_Total_linecut_peaks.txt')
xtot = zcut[:,0:1]
ytot = zcut[:,1:2]
plot1 = ax1.plot(xtot,ytot,color='black')
ax1.set_xlabel('(H,0,1)')
ax1.set_ylabel('Signal')
trans = mtransforms.ScaledTranslation(-20/72, 7/72, fig.dpi_scale_trans)
ax1.text(0.0, 1.0, 'a)', transform=ax1.transAxes + trans,
            fontsize='medium',weight='bold', va='bottom', fontfamily='serif')
ax1.ticklabel_format(axis='y',style='sci',scilimits=(0,0))

zcut = np.loadtxt('normData_MnTe_370K-330K_Total_int_linecut_peaks.txt')
xtot = zcut[:,0:1]
ytot = zcut[:,1:2]
plot2 = ax2.plot(xtot,ytot,color='black')
ax2.set_xlabel('(H,0,1)')
ax2.set_ylabel('Signal')
trans = mtransforms.ScaledTranslation(-20/72, 7/72, fig.dpi_scale_trans)
ax2.text(0.0, 1.0, 'b)', transform=ax2.transAxes + trans,
            fontsize='medium',weight='bold', va='bottom', fontfamily='serif')
ax2.ticklabel_format(axis='y',style='sci',scilimits=(0,0))

plt.tight_layout(pad=4)
plt.savefig('C:/Users/chris/OneDrive/Documents/thesis/figures/normData_MnTe_KAREN_linecuts',dpi=500,bbox_inches = 'tight', pad_inches = .018)
```
